# Algorithm for sort music songs 

## Libraries and dependencies

```python
import pandas as pd
```

## Data Ingestion

```python
path = 'data/spotify_data.csv'
```

```python
# read a csv 
df = pd.read_csv(path)
```

## Preprocessing 

```python
#show me all the columns
df.columns
```

```python
# drop columns album_id, album, track_number, disc_number, explicit, is_local, track_id
df.drop(['album_id', 'album', 'track_number', 'disc_number', 'explicit', 'is_local', 'track_id', 'key', 'time_signature', 'mode' ], axis=1, inplace=True) 
```

```python
df
```

```python
# Convert the album_release_date column to a datetime format
df['track_date'] = pd.to_datetime(df['album_release_date'], format='%Y-%m-%d', errors='coerce')

# Set a default day and month for values that are only in the format YYYY
df['track_date'] = df['track_date'].fillna(pd.to_datetime(df['album_release_date'].astype(str) + '-01-01', format='%Y-%m-%d', errors='coerce'))
```

```python
#drop the album_release_date column and only conserve the year of the track_date column
df.drop(['album_release_date'], axis=1, inplace=True)
df['track_date'] = df['track_date'].dt.year
```
